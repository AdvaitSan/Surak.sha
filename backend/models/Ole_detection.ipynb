```python
from google.colab import files

uploaded = files.upload()
```

```python
from oletools.olevba import VBA_Parser

# Automatically get the filename of the uploaded file
import os
filename = next(iter(uploaded))

# Run olevba analysis
vba_parser = VBA_Parser(filename)

if vba_parser.detect_vba_macros():
    print(f"[*] Macros found in: {filename}\n")
    for (filename, stream_path, vba_filename, vba_code) in vba_parser.extract_macros():
        print(f"--- Macro in {vba_filename} ---\n")
        print(vba_code)

    print("\n[*] Suspicious Indicators:")
    for kw_type, keyword, description in vba_parser.analyze_macros():
        print(f"- {keyword}: {description}")
else:
    print("No macros found.")
```
